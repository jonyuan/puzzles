```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
def simulate_once():
    points = np.random.uniform(-0.5, 0.5, (8, 2))
    max_coords = np.max(np.abs(points), axis=1)
    max_coords_sorted = np.sort(max_coords)
    side_length = 2 * max_coords_sorted[3]
    return side_length, points, max_coords_sorted[3]

def simulate_many(n=100000):
    results = [simulate_once()[0] for _ in range(n)]
    return np.mean(results)
```

```python
side_length, points, half_side = simulate_once()

fig, ax = plt.subplots(figsize=(8, 8))
outer_square = Rectangle((-0.5, -0.5), 1, 1, fill=False, edgecolor='black', linewidth=2)
ax.add_patch(outer_square)
inner_square = Rectangle((-half_side, -half_side), 2*half_side, 2*half_side, 
                         fill=False, edgecolor='red', linewidth=2, linestyle='--')
ax.add_patch(inner_square)
ax.scatter(points[:, 0], points[:, 1], s=100, c='blue', zorder=3)
ax.set_xlim(-0.6, 0.6)
ax.set_ylim(-0.6, 0.6)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title(f'Inner square side length: {side_length:.3f}')
plt.show()
```

```python
avg = simulate_many(100000)
print(f"Simulation average (100k samples): {avg:.6f}")
```
